```python
# runtime, rating, no of rating is obj can be change to int or float for easy analysis
# column gross in us & canada and opening weekend gross in us & canada is missing nearly 40% of total data so dropping them.

removed_df = df[['Gross in US & Canada', 'Opening Weekend Gross in US & Canada']]

df = df.drop(columns=['Gross in US & Canada', 'Opening Weekend Gross in US & Canada'], axis=1)
```

```python
df['Runtime'] = df["Runtime"].str.extract(r'(?:(\d+)h)?\s*(?:(\d+)m)?').astype(float).dot([60,1])
```

```python
df = df.rename(columns={"Runtime": 'runtime_min'})
```

```python
df.columns
```

```python
df['Rating'][5:10]
```

```python
df["Rating"] = df['Rating'].str.extract(r'(\d+)/10').astype(float)
df = df.rename(columns={"Rating": "rating_outof_10"})
```

```python
df['rating_outof_10'].isna().sum()
```

```python
df.head(5)
```

```python
counts = df['Number of Ratings'].str.extract(r'(\d+\.?\d*)([KM]?)')
```

```python
pd.unique(counts[1])
```

```python
df['Number of Ratings'] = (counts[0].astype(float) * counts[1].map({'K': 1000, "M": 1000000, '':1}))
```

```python
df = df.rename(columns={'Number of Ratings': 'no_of_rating'})
```

```python
print(df['no_of_rating'].isnull().sum())
df = df.fillna({'Budget': 0})
```

```python
df['Budget'].isnull().sum()
```

```python
df
```

```python
df['Budget'] = (df['Budget'].str.replace(r'[^\d]', '', regex=True).astype(float))
```

```python
pd.unique(df['Gross worldwide'])
```

```python
df['Gross worldwide'] = (df['Gross worldwide'].str.replace(r'[^\d]', '', regex=True).astype(float))
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['Director'] = df['Director'].fillna('Unknown')
df['Writer'] = df['Writer'].fillna('Unknown')
df['Motion Picture Rating'] = df['Motion Picture Rating'].fillna('Unrated')
df['runtime_min'] = df['runtime_min'].fillna(df['runtime_min'].median())
df['rating_outof_10'] = df['rating_outof_10'].fillna(df['rating_outof_10'].median())
```

```python
df_budget = df.dropna(subset=['Budget', 'Gross worldwide'])
```

```python
import seaborn as sns

import matplotlib.pyplot as plt

# 1. Highest rated movies by genre
print("=" * 60)
print("HIGHEST RATED MOVIES BY GENRE")
print("=" * 60)

for genre in genres:
    top_movies = df[df['Main Genres'].str.contains(genre, case=False, na=False)].nlargest(3, 'rating_outof_10')[['Title', 'rating_outof_10', 'Release Year']]
    print(f"\n{genre}:")
    print(top_movies.to_string(index=False))

# 2. Trends over years
print("\n" + "=" * 60)
print("TRENDS OVER YEARS")
print("=" * 60)

trend_data = df.groupby('Release Year').agg({
    'rating_outof_10': 'mean',
    'no_of_rating': 'mean',
    'Budget': 'mean',
    'Gross worldwide': 'mean'
}).reset_index()

plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
plt.plot(trend_data['Release Year'], trend_data['rating_outof_10'], marker='o')
plt.title('Average Rating Over Years')
plt.xlabel('Year')
plt.ylabel('Average Rating')

plt.subplot(2, 2, 2)
plt.plot(trend_data['Release Year'], trend_data['Budget'], marker='o', color='green')
plt.title('Average Budget Over Years')
plt.xlabel('Year')
plt.ylabel('Budget')

plt.subplot(2, 2, 3)
plt.plot(trend_data['Release Year'], trend_data['Gross worldwide'], marker='o', color='red')
plt.title('Average Gross Worldwide Over Years')
plt.xlabel('Year')
plt.ylabel('Gross')

plt.tight_layout()
plt.show()

# 3. Director and actor analysis
print("\n" + "=" * 60)
print("DIRECTOR ANALYSIS")
print("=" * 60)

top_directors = df.groupby('Director').agg({
    'Title': 'count',
    'rating_outof_10': 'mean'
}).rename(columns={'Title': 'Movie Count'}).sort_values('rating_outof_10', ascending=False).head(10)
print(top_directors)

# 4. Rating vs Revenue
print("\n" + "=" * 60)
print("RATING VS REVENUE ANALYSIS")
print("=" * 60)

plt.figure(figsize=(10, 6))
plt.scatter(df_budget['rating_outof_10'], df_budget['Gross worldwide'], alpha=0.5)
plt.xlabel('Rating (out of 10)')
plt.ylabel('Gross Worldwide Revenue')
plt.title('Rating vs Revenue')
correlation = df_budget['rating_outof_10'].corr(df_budget['Gross worldwide'])
plt.text(0.05, 0.95, f'Correlation: {correlation:.3f}', transform=plt.gca().transAxes, verticalalignment='top', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))
plt.tight_layout()
plt.show()
```
